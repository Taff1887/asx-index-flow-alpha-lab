# 02 - ASX flow-event discovery

**What:** turn holdings snapshots + announcements into the unified events table, and surface where ASX names are being added to (often obscure) global products.

**Why:** the alpha thesis is *forced flow the market hasn't priced*. Step one is simply finding the flow events touching ASX names.

**Economics:** a new position / weight increase in a passive product implies the product *must* buy. If liquidity is thin and the stock hasn't moved, that demand is latent.

```python
import sys; from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / 'src'))
import pandas as pd
from index_flow.config import load_config
from index_flow.event_builder import build_events
cfg = load_config(); cfg.ensure_dirs()
events = build_events(cfg)
print('events:', len(events))
events.head()
```
